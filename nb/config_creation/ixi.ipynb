```python
import json
import os
from glob import glob
```

```python
path = "../data/test_datasets/ixi/"
subjs= os.listdir(path)
dataset = []

for s in subjs:
    d = {"subject_ID" : s}
    
    try:
        d["t2"] = [a.replace(path, "") for a in glob(os.path.join(path, s, "T2*"))][0]
        d["t1"] = [a.replace(path, "") for a in glob(os.path.join(path, s, "T1*"))][0]
        d["mask"] = os.path.join(s, "mask.nii.gz")
    except:
        continue
    
    if os.path.exists(os.path.join(path, s, "mask.nii.gz")):
        dataset.append(d)

    
files = {"training" : [], "validation" : dataset}    
    
# write to json
output_file_path = "../data/test_datasets/ixi_all.json"
with open(output_file_path, "w") as f:
    json.dump(files, f)
print(len(files["validation"]))
```

```python
# filter for images with age available
import pandas as pd

meta = "/path/to/IXI/metadata_ixi_institutes.csv"

df = pd.read_csv(meta)
df
```

```python
df.columns
```

```python
# stratify on age, sex, institute

from sklearn.model_selection import train_test_split

strat_df = df[["IXI_ID", "AGE", "institute", "SEX_ID (1=m, 2=f)"]].dropna()
# cut age into bins of 10 years
strat_df["AGE"] = pd.cut(strat_df["AGE"], bins=[18, 30, 40, 50, 60, 70, 100])
strat_df["a"] = strat_df["AGE"].astype(str) + "_" + strat_df["institute"].astype(str) + "_" + strat_df["SEX_ID (1=m, 2=f)"].astype(str)
strat_df["a"] = strat_df.a.str.replace("(70, 100]_IOP_1", "(60, 70]_IOP_1")

train, test = train_test_split(strat_df, test_size=0.2, stratify=strat_df["a"], random_state=42)
```

```python
path = "../data/test_datasets/ixi/"
subjs= os.listdir(path)
dataset = []

for s in subjs:
    d = {"subject_ID" : s}
    
    try:
        d["t2"] = [a.replace(path, "") for a in glob(os.path.join(path, s, "T2*"))][0]
        d["t1"] = [a.replace(path, "") for a in glob(os.path.join(path, s, "T1*"))][0]
        d["pd"] = [a.replace(path, "") for a in glob(os.path.join(path, s, "PD*"))][0]
        d["mask"] = os.path.join(s, "mask.nii.gz")
    except:
        continue
    
    if os.path.exists(os.path.join(path, s, "mask.nii.gz")):
        dataset.append(d)

train_cases, test_cases = [], []

for t in dataset:
    ixi_id = int(t["subject_ID"].replace("IXI", ""))
    if ixi_id in train["IXI_ID"].values:
        train_cases.append(t)
    elif ixi_id in test["IXI_ID"].values:
        test_cases.append(t)

len(train_cases), len(test_cases)
files = {"training" : train_cases, "validation" : test_cases}

# write to json
output_file_path = "../data/test_datasets/ixi_train.json"
with open(output_file_path, "w") as f:
    json.dump(files, f)
```
